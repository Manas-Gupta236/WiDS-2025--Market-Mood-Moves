```python
import yfinance as yf

def download_stock_data(ticker="AAPL", start="2024-01-01", end="2024-12-31"):
    """
    Downloads historical OHLC stock data using yfinance
    """
    data = yf.download(ticker, start=start, end=end)
    return data

if __name__ == "__main__":
    df = download_stock_data()
    print(df.head())
```
