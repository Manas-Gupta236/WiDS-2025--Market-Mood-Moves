Function to calculate the average of a list of stock prices

```python
def average_price(prices):
    return sum(prices) / len(prices)


prices = [145, 150, 152, 148, 149]
print("Average price:", average_price(prices))
```

Classify stocks as expensive (>150) or affordable (<150)

```python
def classify_stock(price):
    if price > 150:
        return "Expensive"
    else:
        return "Affordable"


print(classify_stock(160))
print(classify_stock(140))
```

Loop to compute daily returns over a 10-day period

```python
prices = [100, 102, 101, 103, 105, 104, 106, 108, 107, 109]

daily_returns = []

for i in range(1, len(prices)):
    daily_returns.append(prices[i] - prices[i - 1])

print("Daily returns:", daily_returns)
```

Load a stock price CSV file

```python
import pandas as pd

df = pd.read_csv("stock_prices.csv")
print(df.head())
```

Filter for days where sentiment exceeds 0.7

```python
filtered_df = df[df["sentiment"] > 0.7]
print(filtered_df)
```

Create daily returns column

```python
df["Daily_Return"] = df["Close"] - df["Close"].shift(1)
print(df.head())
```

Calculate weekly average closing price

```python
df["Date"] = pd.to_datetime(df["Date"])
df.set_index("Date", inplace=True)

weekly_avg = df["Close"].resample("W").mean()
print(weekly_avg)
```

Create an array of 10 stock prices & calculate statistics

```python
import numpy as np

prices = np.array([100, 102, 101, 103, 105, 104, 106, 108, 107, 109])

print("Mean:", np.mean(prices))
print("Standard Deviation:", np.std(prices))
print("Minimum:", np.min(prices))
print("Maximum:", np.max(prices))
```
